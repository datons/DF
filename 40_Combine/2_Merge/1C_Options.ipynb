# Merge DataFrames

## Load datasets

```python
import pandas as pd

path = '../../../data/topics/energy/generation/spain/full/wide_ts.csv'
df_generation = pd.read_csv(path)

path = '../../../data/topics/energy/price/spain/spot_merge.csv'
df_price = pd.read_csv(path)
```

## Merge DataFrames

```python
df = pd.merge(df_generation, df_price, on='datetime')
df
```

## Multivariate analysis

For example, a correlation heatmap with seaborn.

```python
df = df.set_index('datetime')
df
```

```python
dfc = df.corr().mul(100)
dfc
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(dfc, annot=True, fmt=".0f", cmap='coolwarm', center=0)
```

```python
columns = ['Saldo Marruecos', 'Turbinación bombeo', 'Carbón']
dfr = df[columns].corr()
dfr
```

```python
import plotly.express as px
px.scatter_matrix(df[columns])
```

## Merge types

![](src/Miro%20Medium%20Image.png)

### Left

```python
df_generation.merge(df_price, on='datetime', how='left')
```

### Right

```python
df_generation.merge(df_price, on='datetime', how='right')
```

### Outer

```python
df_generation.merge(df_price, on='datetime', how='outer')
```

### Inner

```python
df_generation.merge(df_price, on='datetime', how='inner')
```
